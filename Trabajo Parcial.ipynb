# Descarga de archivo:

## LEYENDO EL ARCHIVO 

```python

file_path = "gene.fna"
current_id = ""
current_description = ""
current_sequence = ""

with open(file_path, "r") as file:
    for line in file:
        line = line.strip() 
        
        if line.startswith(">"):  
            
            if current_id:
                print(f"ID: {current_id}")
                print(f"Descripci√≥n: {current_description}")
                print(f"Secuencia: {current_sequence}\n")
            
            current_id, current_description = line[1:].split(maxsplit=1)
            current_sequence = ""  # Reiniciar la secuencia para el nuevo registro
        else:
            current_sequence += line  # Concatenar la l√≠nea a la secuencia actual
if current_id:
    print(f"ID: {current_id}")
    print(f"Descripci√≥n: {current_description}")
    print(f"Secuencia: {current_sequence}\n")
```

### PREGUNTA 1

```python
# Secuencia de ADN proporcionada
import re
count_C = len(re.findall(r'C', current_sequence))

# Longitud de la secuencia
total_length = len(current_sequence)

# Calcular porcentaje de "C":
porcentaje_C = (count_C  / total_length) * 100

print(f"El porcentaje de nucle√≥tidos 'C' en la secuencia es: {porcentaje_C:.2f}%")
```

```python
count_G = len(re.findall(r'G', current_sequence))

# Longitud de la secuencia
total_length = len(current_sequence)

# Calcular porcentaje de "G" :
porcentaje_G = (count_G  / total_length * 100)

print(f"El porcentaje de nucle√≥tidos 'G' en la secuencia es: {porcentaje_G:.2f}%")
```

## PREGUNTA 2

```python
import re

secuencia = current_sequence

contadorccg = len(re.findall(r"(?=(CCG))", secuencia))

contador_gcc = len(re.findall(r"(?=(GCC))", secuencia))

print(f"N√∫mero de GCC: {contador_gcc}")
print(f"N√∫mero de CCG: {contadorccg}")

total_contador= contador_gcc + contadorccg 

print(f"N√∫mero total de CCG o GCC que se encuentran en la secuencia: {total_contador}")

```
