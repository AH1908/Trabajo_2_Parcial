# Descarga de archivo:

## LEYENDO EL ARCHIVO 

```python

file_path = "gene.fna"
current_id = ""
current_description = ""
current_sequence = ""

with open(file_path, "r") as file:
    for line in file:
        line = line.strip() 
        
        if line.startswith(">"):  
            
            if current_id:
                print(f"ID: {current_id}")
                print(f"Descripci√≥n: {current_description}")
                print(f"Secuencia: {current_sequence}\n")
            
            current_id, current_description = line[1:].split(maxsplit=1)
            current_sequence = ""  # Reiniciar la secuencia para el nuevo registro
        else:
            current_sequence += line  # Concatenar la l√≠nea a la secuencia actual
if current_id:
    print(f"ID: {current_id}")
    print(f"Descripci√≥n: {current_description}")
    print(f"Secuencia: {current_sequence}\n")
```
