# Descarga de archivo:

## LEYENDO EL ARCHIVO 

```python
import re

with open('gene.fna', 'r') as file:
    data = file.read()

secuencias= ''.join(data.split('\n')[1:])
print(secuencias)
```

### PREGUNTA 1

```python
import re


# Contar nucleótidos 'C'
count_C = len(re.findall(r'C', secuencias))

# Contar nucleótidos 'G'
count_G = len(re.findall(r'G', secuencias))

# Longitud de la secuencia
total_length = len(secuencias)

# Calcular porcentaje de "C"
porcentaje_C = (count_C / total_length) * 100
print(f"El porcentaje de nucleótidos 'C' en la secuencia es: {porcentaje_C:.2f}%")

# Calcular porcentaje de "G"
porcentaje_G = (count_G / total_length) * 100
print(f"El porcentaje de nucleótidos 'G' en la secuencia es: {porcentaje_G:.2f}%")

# Calcular el porcentaje total de "G" y "C"
porcentaje_total = porcentaje_C + porcentaje_G
print(f"El porcentaje de nucleótidos 'G' o 'C' en la secuencia es: {porcentaje_total:.2f}%")
```

## PREGUNTA 2

```python
import re


contadorccg = len(re.findall(r"(?=(CCG))", secuencias))

contador_gcc = len(re.findall(r"(?=(GCC))", secuencias))

print(f"Número de GCC: {contador_gcc}")
print(f"Número de CCG: {contadorccg}")

total_contador= contador_gcc + contadorccg 

print(f"Número total de CCG o GCC que se encuentran en la secuencia: {total_contador}")

```

## PREGUNTA 3

```python
conteo = secuencias.count('GAATTC')

print(f'La secuencia "GAATTC" se encuentra {conteo} veces en la cadena de ADN.')
```
